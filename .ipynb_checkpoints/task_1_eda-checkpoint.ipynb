```python
# Cell 1: Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style for plots for better aesthetics
sns.set_style("whitegrid")

# This is a magic command specific to Jupyter notebooks
# It ensures that plots are displayed directly in the notebook output
%matplotlib inline

print("Libraries imported successfully!")
```

```python
# Cell 2: Load the Data (Corrected for Pipe Delimiter)

# Define the file path. Adjust if your file is not in a 'data' folder
file_path = 'data/MachineLearningRating_v3.txt'

try:
    # Attempt to read with pipe '|' as separator
    df = pd.read_csv(file_path, sep='|')
    print("Data loaded successfully with pipe '|' delimiter. Displaying first 5 rows:")
    print(df.head())
    print("\nDataFrame Info:")
    df.info()

    # After successful loading, it's a good practice to strip whitespace from column names
    # just in case there are subtle leading/trailing spaces from the source file.
    df.columns = df.columns.str.strip()
    print("\nColumns after stripping whitespace:")
    print(df.columns.tolist())

except Exception as e:
    print(f"Error loading with pipe delimiter: {e}")
    print("Please double-check your .txt file's actual delimiter.")

# Quick check on shape
if 'df' in locals(): # Check if df was successfully created
    print(f"\nDataFrame shape: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# Cell 3: Data Structure - Initial Review and Type Conversion (REVISED)

print("Initial data types (dtypes):")
print(df.dtypes)

# --- REVISED PART START ---
# PROBLEM: 'TransactionDate' column is not in the DataFrame based on your dtypes output.
# SOLUTION: We will assume 'TransactionMonth' (which is 'object' type) actually holds the date string,
# and we will rename it to 'TransactionDate' for clarity, then convert it.

# First, confirm if 'TransactionMonth' contains full date strings or just month names.
# Print a few unique values from df['TransactionMonth'] to verify.
print("\nFirst 10 unique values from 'TransactionMonth' column:")
print(df['TransactionMonth'].head(10).tolist())
# If these look like full dates (e.g., '2014-02-01', '01-FEB-2014'), proceed.
# If they are just 'Feb', 'Mar', etc., then the full date column is truly missing,
# or it has a different name than listed in the assignment, or it's within another column.

# If 'TransactionMonth' contains date strings, rename it and convert:
if 'TransactionMonth' in df.columns and df['TransactionMonth'].dtype == 'object':
    # This assumes 'TransactionMonth' column, which is an object type, is actually your date column.
    # We rename it to 'TransactionDate' for consistency with the assignment's implied name.
    # If the original 'TransactionMonth' truly contained only month names (like 'Feb'),
    # then this step is incorrect and implies the actual date column is missing from your raw data.
    df.rename(columns={'TransactionMonth': 'TransactionDate'}, inplace=True)
    print("\nRenamed 'TransactionMonth' to 'TransactionDate'.")

    # Now convert the renamed 'TransactionDate' column to datetime objects
    df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], errors='coerce')

    # Now, if we need a *separate* column for just the month name (e.g., 'Feb', 'Mar'), we can create it from TransactionDate
    # Let's call it 'MonthName' to distinguish it, or 'TransactionMonthName'
    df['TransactionMonthName'] = df['TransactionDate'].dt.strftime('%b') # e.g., 'Feb', 'Mar'
    print("Created 'TransactionMonthName' from 'TransactionDate'.")

else:
    print("\nError: 'TransactionMonth' column not found or not of 'object' dtype as expected for date conversion.")
    print("Please check your data's columns carefully. The 'TransactionDate' column seems to be missing.")
    print("DataFrame columns currently are:", df.columns.tolist())


print("\nData types after (attempted) converting 'TransactionDate' and adding 'TransactionMonthName':")
print(df.dtypes)
print("\nFirst 5 rows with updated date columns:")
print(df[['TransactionDate', 'TransactionMonthName']].head()) # Now checking the new name if created

# Review unique values for some categorical columns to spot inconsistencies
print("\nUnique values for 'Gender':", df['Gender'].unique())
print("Unique values for 'MaritalStatus':", df['MaritalStatus'].unique())
print("Unique values for 'Province':", df['Province'].unique())


# It's good practice to convert object columns that are truly categorical to 'category' dtype
# This can save memory and speed up some operations later
# Make sure to re-check after any renaming
for col in df.select_dtypes(include='object').columns:
    # Avoid converting the date column if it was not successfully converted to datetime
    if col != 'TransactionDate': # Only convert if it's still an object and not our target date column
        df[col] = df[col].astype('category')

print("\nData types after converting remaining object columns to 'category':")
print(df.dtypes)
```

```python
# Cell 4: Data Quality Assessment - Missing Values

print("Checking for missing values:")
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_info = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage (%)': missing_percentage
})

# Filter to show only columns with missing values
missing_info = missing_info[missing_info['Missing Count'] > 0].sort_values(by='Missing Percentage (%)', ascending=False)

if not missing_info.empty:
    print(missing_info)
else:
    print("No missing values found in the dataset.")

print("\nDecisions about missing values (for later modeling, just noting for now):")
print("- Columns with high missing percentage might need to be dropped or carefully imputed.")
print("- Numerical columns might be imputed with mean/median.")
print("- Categorical columns might be imputed with mode or 'Unknown'.")
```

```python
# Cell 5: Data Summarization - Descriptive Statistics for Numerical Features

print("Descriptive Statistics for Numerical Features:")
# Select only numerical columns for descriptive statistics
numerical_cols = df.select_dtypes(include=np.number).columns
print(df[numerical_cols].describe().T) # .T transposes for better readability

print("\nVariability Analysis (Standard Deviation and Range):")
for col in numerical_cols:
    std_dev = df[col].std()
    data_range = df[col].max() - df[col].min()
    print(f"- {col}: Standard Deviation = {std_dev:.2f}, Range = {data_range:.2f}")

print("\nObservations on variability:")
print("- High standard deviation relative to the mean suggests high variability.")
print("- Large ranges indicate the spread of data from min to max.")
print("- For instance, 'TotalPremium' and 'TotalClaims' likely show high variability, indicating diverse policy values and claim amounts.")
```

```python
# Cell 6: Univariate Analysis - Distribution of Numerical Variables

# Key financial variables
financial_vars = ['TotalPremium', 'TotalClaims', 'CustomValueEstimate', 'SumInsured', 'CalculatedPremiumPerTerm', 'ExcessSelected']

plt.figure(figsize=(18, 12))
for i, col in enumerate(financial_vars):
    plt.subplot(2, 3, i + 1) # 2 rows, 3 columns
    sns.histplot(df[col].dropna(), kde=True, bins=50) # dropna to handle potential NaNs
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

print("\nObservations from Histograms:")
print("- Many financial variables often show a skewed distribution (e.g., right-skewed for premiums/claims), with a few very high values.")
print("- This indicates the presence of outliers or a long tail for higher amounts.")

# Box plots for outlier detection
plt.figure(figsize=(18, 8))
for i, col in enumerate(financial_vars):
    plt.subplot(1, 6, i + 1)
    sns.boxplot(y=df[col].dropna())
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()

print("\nObservations from Box Plots (Outlier Detection):")
print("- The box plots clearly show numerous outliers, especially for 'TotalClaims', 'TotalPremium', and 'CustomValueEstimate'.")
print("- These outliers represent policies with exceptionally high values or claims.")
print("- For modeling, these outliers may need special handling (e.g., transformation, capping, or robust models).")
```

```python
# Cell 7: Univariate Analysis - Distribution of Categorical Variables

# Select a few key categorical columns for visualization
categorical_vars = ['Province', 'VehicleType', 'Gender', 'MaritalStatus', 'Product', 'CoverGroup', 'Bank']

plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_vars):
    plt.subplot(3, 3, i + 1) # Adjust subplot grid as needed
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel('') # Remove y-label as it's redundant with tick labels
plt.tight_layout()
plt.show()

print("\nObservations from Bar Charts:")
print("- 'Province': Gauteng likely has the highest number of policies, followed by Western Cape, indicating major markets.")
print("- 'VehicleType': Cars are probably the dominant vehicle type.")
print("- 'Gender': Observe the distribution of policies between Male and Female, which is important for hypothesis testing.")
print("- 'Product' and 'CoverGroup': Show popular insurance products/coverages.")
print("- 'Bank': Reveals the distribution of clients across different banks.")
```

```python
# Cell 8: Bivariate/Multivariate Analysis - Overall Loss Ratio and Variation

# Calculate the overall Loss Ratio
overall_total_claims = df['TotalClaims'].sum()
overall_total_premium = df['TotalPremium'].sum()

# Add a small epsilon to avoid division by zero if TotalPremium is ever 0
epsilon = 1e-6
overall_loss_ratio = overall_total_claims / (overall_total_premium + epsilon)
print(f"Overall Loss Ratio (TotalClaims / TotalPremium): {overall_loss_ratio:.4f}\n")

# How Loss Ratio varies by Province, VehicleType, and Gender
risk_factors = ['Province', 'VehicleType', 'Gender']

plt.figure(figsize=(18, 6))
for i, factor in enumerate(risk_factors):
    # Group by the factor and calculate sum of claims and premium
    grouped_data = df.groupby(factor).agg(
        TotalClaims=('TotalClaims', 'sum'),
        TotalPremium=('TotalPremium', 'sum')
    ).reset_index()

    # Calculate Loss Ratio for each group, handling potential division by zero
    grouped_data['LossRatio'] = grouped_data['TotalClaims'] / (grouped_data['TotalPremium'] + epsilon)

    plt.subplot(1, 3, i + 1)
    sns.barplot(data=grouped_data.sort_values(by='LossRatio', ascending=False),
                x='LossRatio', y=factor, palette='coolwarm')
    plt.title(f'Loss Ratio by {factor}')
    plt.xlabel('Loss Ratio')
    plt.ylabel(factor)
plt.tight_layout()
plt.show()

print("\nObservations on Loss Ratio Variation:")
print("- 'Province': Reveals which provinces have higher or lower claim costs relative to premiums. This is directly relevant to Hypothesis 1.")
print("- 'VehicleType': Shows if certain vehicle types are riskier.")
print("- 'Gender': Indicates if there's a significant difference in risk profile between men and women (relevant to Hypothesis 4).")
print("- These initial insights will guide our formal hypothesis testing.")
```

```python
# Cell 9: Bivariate Analysis - Correlations and Temporal Trends 


correlation_cols = ['TotalPremium', 'TotalClaims', 'CustomValueEstimate', 'SumInsured',
                    'CalculatedPremiumPerTerm', 'ExcessSelected',
                    'kilowatts', # Changed from 'Kilowatts'
                    'cubiccapacity', # Changed from 'Cubiccapacity'
                    'NumberOfDoors', 'CapitalOutstanding', 'NumberOfVehiclesInFleet']

actual_numerical_cols = [col for col in correlation_cols if pd.api.types.is_numeric_dtype(df[col])]

print(f"Columns used for correlation matrix: {actual_numerical_cols}")

# Calculate correlation matrix
# Ensure df[actual_numerical_cols] does not contain any object types that cannot be correlated
# .corr() will automatically drop non-numeric columns it can't handle, but explicitly checking is better
corr_matrix = df[actual_numerical_cols].corr()


plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Key Numerical Features')
plt.show()

print("\nObservations from Correlation Matrix:")
print("- Look for strong positive or negative correlations. For example:")
print("  - 'TotalPremium' and 'SumInsured' are likely highly correlated.")
print("  - 'TotalClaims' might have moderate correlations with 'TotalPremium' or 'CustomValueEstimate'.")
print("  - Very high correlations between independent variables can indicate multicollinearity, which is a consideration for some models.")

# Temporal Trends: Claim frequency/severity over the 18-month period
# Ensure 'TransactionDate' is datetime and 'TransactionMonthName' is created from Cell 3
# The previous problem was that TransactionDate was not found. Now it should be,
# based on our fix in Cell 3 (renaming 'TransactionMonth' to 'TransactionDate')

# Filter out NaT values from TransactionDate before grouping
df_filtered_dates = df.dropna(subset=['TransactionDate']).copy()

# Calculate Claim Occurred for frequency
df_filtered_dates['HasClaim'] = (df_filtered_dates['TotalClaims'] > 0).astype(int)

# Use 'TransactionDate' for grouping by month and year
monthly_trends = df_filtered_dates.groupby(df_filtered_dates['TransactionDate'].dt.to_period('M')).agg(
    TotalClaims=('TotalClaims', 'sum'),
    TotalPremium=('TotalPremium', 'sum'),
    NumPolicies=('PolicyID', 'nunique'), # Count unique policies for a more accurate count
    ClaimFrequency=('HasClaim', 'mean'), # Mean of 1s and 0s gives frequency
    NumClaims=('HasClaim', 'sum') # Count of claims
).reset_index()


# Calculate Claim Severity: Total Claims / Number of Claims (only when claims occurred)
# Avoid division by zero: if NumClaims is 0, severity is 0 or NaN
monthly_trends['ClaimSeverity'] = monthly_trends.apply(
    lambda row: row['TotalClaims'] / row['NumClaims'] if row['NumClaims'] > 0 else np.nan, # Use np.nan for no claims
    axis=1
)

# Convert Period back to datetime for plotting ease
monthly_trends['TransactionMonthYear'] = monthly_trends['TransactionDate'].dt.to_timestamp() # Use the TransactionDate column directly

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.lineplot(data=monthly_trends, x='TransactionMonthYear', y='ClaimFrequency', marker='o')
plt.title('Claim Frequency Over Time')
plt.xlabel('Month')
plt.ylabel('Claim Frequency (Proportion of policies with claims)')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.lineplot(data=monthly_trends.dropna(subset=['ClaimSeverity']), x='TransactionMonthYear', y='ClaimSeverity', marker='o') # dropna to not plot NaNs
plt.title('Claim Severity Over Time')
plt.xlabel('Month')
plt.ylabel('Average Claim Amount (for policies with claims)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

print("\nObservations on Temporal Trends:")
print("- 'Claim Frequency': Shows if the proportion of policies making claims changed over the 18 months.")
print("- 'Claim Severity': Indicates if the average cost per claim increased or decreased over time.")
print("- Look for any seasonal patterns or significant spikes/drops that could correlate with external events.")
```

```python
# Cell 10: Creative and Beautiful Plots

loss_ratio_by_province_cover = df.groupby(['Province', 'CoverGroup']).agg(
    TotalClaims=('TotalClaims', 'sum'),
    TotalPremium=('TotalPremium', 'sum')
).reset_index()

epsilon = 1e-6 # To prevent division by zero
loss_ratio_by_province_cover['LossRatio'] = loss_ratio_by_province_cover['TotalClaims'] / (loss_ratio_by_province_cover['TotalPremium'] + epsilon)

# Pivot the table for heatmap
pivot_table = loss_ratio_by_province_cover.pivot_table(index='Province', columns='CoverGroup', values='LossRatio')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu", linewidths=.5)
plt.title('Loss Ratio Heatmap: Province vs. Cover Group')
plt.xlabel('Cover Group')
plt.ylabel('Province')
plt.show()
print("Insight from Plot 1: This heatmap highlights specific combinations of provinces and cover groups that have particularly high or low loss ratios, indicating areas of high or low risk, respectively. This can directly inform pricing strategies for specific products in specific regions.")



plt.figure(figsize=(12, 7))
sns.violinplot(data=df, x='VehicleType', y='NumberOfVehiclesInFleet', palette='pastel')
plt.title('Distribution of Fleet Size by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Vehicles in Fleet')
plt.show()
print("\nInsight from Plot 2: This violin plot shows the typical fleet size for different vehicle types. For example, it might reveal that 'Car' policies are usually for single vehicles, while 'Truck' or 'Commercial Vehicle' policies are often associated with larger fleets. This informs understanding of client segmentation and potential bundled product offerings.")



plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Gender', y='TotalPremium', palette='Paired')
plt.title('Distribution of Total Premium by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Premium')

plt.subplot(1, 2, 2)

sns.boxplot(data=df[df['TotalClaims'] > 0], x='Gender', y='TotalClaims', palette='Paired')
plt.title('Distribution of Total Claims (for claims > 0) by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Claims Amount')

plt.tight_layout()
plt.show()
print("\nInsight from Plot 3: These box plots visually compare the distribution of premiums paid and claim amounts between different genders. While boxes show central tendency and spread, the outliers are also visible. This visual evidence will be statistically validated in Task 3 for the hypothesis regarding risk differences between genders.")
```
